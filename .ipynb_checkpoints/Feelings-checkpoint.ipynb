# [Signate Beginner Competition](https://signate.jp/competitions/463)

**以下称号限定コンペの説明（引用）**

SIGNATEにはコンペの戦績に応じて6つの称号（Tier）が設定されています。<br>
中でもBiginnerはSIGNATEへの会員登録により付与される称号で、１つ上の称号であるIntermidiateに昇格するためには、最低１つのコンペにおいて上位60%以上の成績をおさめる必要があります。<br>
参加可能なタイミングで開催されているコンペは難易度も様々、参加者レベルや人数もまちまちなので、はじめたばかりの方には少し取り組みにくいかもしれません。
そこで、自分のペースで気軽に参加可能なBeginner限定コンペを設定しました。<br>
特徴は以下となります。<br>

* 参加資格はBeginnerのみ
* 賞金はなし
* 期間は１ヶ月間
* 毎月開催、テーマはテーブル・画像データなど多様で比較的簡単なタスク
* あらかじめ設定されたスコアを上回った段階で、即座にIntermidiateへ昇格

参加者同士で競い合うのではなく、自分の好みのテーマに好きなタイミングで取り組むことが可能です。コンペに慣れ、次へのステップアップの機会としてご利用ください。

## data
医療保険の費用帯予測

年齢、性別、BMI、子供の数、喫煙、地域から価格帯（低、中、高）の分類を行う<br>
レコード数：1600レコード

## 評価
評価関数「F1score」（マクロ平均）<br>
運営の設定するTier変更の閾値:F1macro=0.804

## 結果
F1score=0.8127132, 13/129位（6月24日現在）を達成し、BeginnerからIntermidiateに昇格できた。

## 行ったことと得られたこと

- データの可視化
    - Pandasを用いて各特徴量の統計量や相関係数、欠損値の確認をした
    - 今回のデータには欠損値は無く、また特徴量も6つとシンプルなデータであった
- 探索的データ解析
    - 闇雲にモデルにデータを与えてもナンセンス
    - 仮説をたて、箱ひげ図や積み上げ棒グラフ等を作成して確かめた
- 機械学習モデルの選定
    - 研修で扱ったモデルをいくつか試した
        - ロジスティクス回帰/主成分分析/ランダムフォレスト/lightGBM
    - 特徴量をいじらない場合ロジスティクス回帰が最もF1値が高かった
    - ダウンサンプリングやアンサンブル学習なども行った
        - そもそものデータの量が小さいためかダウンサンプリングは効果的ではなかった
- 特徴量エンジニアリング
    - 集計した新たな特徴量を作成し、最終的に6個から49個まで増やした
    - lightGBMで寄与率を確認したが新たに作った特徴量で特別効いたものはなかったが、年齢層と性別ごとに集計したBMIは比較的効いていた
    - 今回試せなかった集計の方法や新たな特徴量の作り方などをもっと知る必要がある
- パラメータ調整
    - Grid Searchやoptunaを使ってパラメータ調整を行った
    - ハイパーパラメータの意味を通して機械学習モデルについてすることができた
        - lightGBMには正則化項が複数あることなど

## 所感
- 機械学習による分析の一連の流れを把握できた
- 結局もっとも効いたのは特徴量のを絞ることだった
- EDAをしっかりすることで必要な特徴量がわかるのでEDAをスマートに行う術を身につける
- ランダムフォレストなどを使って変数重要度を確認しながら実験を行うと良さそう
- 特徴量の多いデータになると扱いが難しくなりそう
- データに対するドメイン知識が必須
